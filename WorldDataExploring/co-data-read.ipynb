```python
import numpy as np
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt
%matplotlib inline


data_series = read_csv('covid_19_data.csv', header=0, parse_dates=['ObservationDate'])
```

```python
data_series
```

```python
dataframe = pd.DataFrame(data_series)
```

```python
countries_data = dataframe.filter(['ObservationDate', 'Country/Region', 'Confirmed'])
```

```python
countries_data_pl = countries_data[ countries_data['Country/Region'] == 'Poland']
countries_data_it = countries_data[ countries_data['Country/Region'] == 'Italy']
```

```python
countries_data_pl
```

```python
countries_data_it
```

```python

italy_confirmed = countries_data_it.filter(['Confirmed'])
italy_confirmed.to_csv('italy_confirmed')
```
